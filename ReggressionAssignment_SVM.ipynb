```python
# import pandas 
import pandas as pd
```

```python
# read insurance data from file
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
independent
```

```python
dependent=dataset[["charges"]]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
final_df = pd.DataFrame(columns=['Algorithim','Kernel','C','R-Score'])
```

```python
from sklearn.svm import SVR
from sklearn.metrics import r2_score
def supportVectorMachine_rscore(ker,c_value,X_train,X_test,y_train,y_test): 
    regressor=SVR(kernel=ker,C=c_value)
    regressor.fit(X_train,y_train)
    y_pred=regressor.predict(X_test)
    r_score=r2_score(y_test,y_pred)
    return r_score
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
algorithim_name =  "Support Vector Machine"
ker_list = ['linear','rbf','sigmoid','poly']
c_value_list = [10,100,500,1000,2000,5000,10000]
for ker in ker_list:
    for c_value in c_value_list:        
            X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
            r_scr = supportVectorMachine_rscore(ker,c_value,X_train,X_test,y_train,y_test)
            data_to_store = {'Algorithim' : algorithim_name,'Kernel' : ker ,'C' : c_value ,'R-Score' : r_scr}
            final_df.loc[len(final_df)] = data_to_store 
```
