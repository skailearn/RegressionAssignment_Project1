```python
# import pandas 
import pandas as pd
```

```python
# read insurance data from file
dataset=pd.read_csv("insurance_pre.csv")
```

```python
#dataset details
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
independent
```

```python
dependent=dataset[["charges"]]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
def randomForestRegressor_rscore(crt,est,feat,X_train,X_test,y_train,y_test):    
    regressor=RandomForestRegressor(criterion=crt,max_features = feat,n_estimators = est, random_state = 0)
    regressor=regressor.fit(X_train,y_train)
    y_pred=regressor.predict(X_test)   
    r_score=r2_score(y_test,y_pred)
    return r_score
```

```python
final_df = pd.DataFrame(columns=['Algorithim','Criterion','Splitter/Estimators','Max_features','R-Score'])
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
algorithim_name =  "RandomForestRegressor"
criterion_list = ['squared_error','friedman_mse','absolute_error','poisson']
n_estimators = [10,100]
max_features_list = ['sqrt','log2',None]
for crt in criterion_list:
    for est in n_estimators:
        for feat in max_features_list:
            X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
            r_scr = randomForestRegressor_rscore(crt,est,feat,X_train,X_test,y_train,y_test)
            data_to_store = {'Algorithim' : algorithim_name,'Criterion' : crt ,'Splitter/Estimators' : est ,'Max_features' : feat ,'R-Score' : r_scr}
            final_df.loc[len(final_df)] = data_to_store 
```

```python
# Best model in decision tree is squared_error,best,lo2 is 0.80
```

```python
final_df.sort_values('R-Score',ascending = [False])
```

```python
final_df.to_csv("random_foreset_r2score_.csv")
```
